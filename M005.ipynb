### Ãœber Matplotlib selbst

```python
plt.figure(figsize=(12, 3))
plt.plot(data["Open"])
plt.title("Open Price")
plt.xlabel("Year")
plt.ylabel("Price ($)")
plt.show()
```

### Globaler vs. Objektorientiert Modus

Beim Globalen Modus wird der Graph Ã¼ber die plt. Methoden konfiguriert und im Anschluss angezeigt

Beim Objektorientierten Modus haben wir Zugriff auf die einzelnen Plots & Achsen und kÃ¶nnen diese separat konfigurieren

#### Beispiel Globaler Modus

```python
plt.figure(figsize=(12, 3))
plt.plot(data["Open"])
plt.title("Open Price")
plt.xlabel("Year")
plt.ylabel("Price ($)")
plt.show()  # Plot beenden


plt.figure(figsize=(12, 3))
plt.plot(data["Close"])
plt.title("Close Price")
plt.xlabel("Year")
plt.ylabel("Price ($)")
```

#### Beispiel Objektorientierter Modus

```python
figure, ((plot1_1, plot1_2), (plot2_1, plot2_2)) = plt.subplots(2, 2, figsize=(12, 4))  # RÃ¼ckgabewert: Figure selbst (das Ganze) + die 4 Plots separat

plot1_1.plot(np.random.random(100), color="Red")
plot1_1.set_title("Rote Linie")

plot1_2.plot(np.random.random(100), color="Green")
plot1_2.set_title("GrÃ¼ne Linie")

plot2_1.plot(np.random.random(100), color="Blue")
plot2_1.set_title("Blue Linie")

plot2_2.plot(np.random.random(100), color="Yellow")
plot2_2.set_title("Gelbe Linie")
```

#### Histogramm

```python
plt.hist(data["Med.Age"], np.arange(15, 50), edgecolor="black", color="yellow")
None
```

#### Tortendiagramm

= Kreisdiagramm

```python
plt.figure(figsize=(10, 10))
plt.pie(data.loc[:5, "WorldShare"], labels=data.loc[:5, "Country"], rotatelabels=True, labeldistance=0.4, wedgeprops={"linewidth": 1, "edgecolor": "black"})
plt.savefig("Test.png")
None
```

#### Streudiagramm

Zeigt die Verteilung von Punkten an

```python
plt.figure(figsize=(12, 6))
points = np.random.randint(10000, size=(100, 100))
plt.scatter(points[0], points[1], c=points[0] * points[1], s=(points[0] * points[1]) / 50000)
None
```
